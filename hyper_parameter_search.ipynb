```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
results = pd.read_csv('results.csv')
```

```python
results = results[results['Iterations'] == 200]
```

## Invloed van population size

```python
pop = results.groupby('Population_size').mean()
plt.plot(pop[['Best_objective', 'Mean_objective']]);
```

## Invloed van de offspring size

```python
off = results.groupby('Offspring_size').mean()
plt.plot(off[['Best_objective', 'Mean_objective']]);
```

## Invloed van k

```python
k = results.groupby('k').mean()
plt.plot(k[['Best_objective', 'Mean_objective']]);
```

## Invloed van alpha

```python
a = results.groupby('alpha').mean()
plt.plot(a[['Best_objective', 'Mean_objective']]);
```

## Beste waarden

```python
bestobjs = results[results['Best_objective'] == results['Best_objective'].min()]
bestobjs
```

```python
meanobjs = results[results['Mean_objective'] == results['Mean_objective'].min()]
meanobjs
```
